```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

Benar_44 = pd.read_csv("https://www.dropbox.com/s/fy4oxmciv4pkylj/Benar-44.csv?dl=1", sep = ";")
Benar_44
```

```python
DT = Benar_44['DT']
DTS = Benar_44["DTS"]
VP = 1000000/DT
VS = 1000000/DTS
Poisson = (VP**2 -2*VS**2)/(2*(VP**2-VS**2)) 
Benar_44['DTS'] = DTS
Benar_44['VP'] = VP
Benar_44['VS'] = VS
Benar_44['Poisson'] = Poisson
Benar_44
```

```python
Benar_44 
Benar_44.to_csv('Benar_44', index = False, sep = ";")
```

```python
Top_ShaleA= Benar_44.loc[3298:3470, :]
Top_ShaleA = Top_ShaleA[Top_ShaleA['VP']>0]
Top_ShaleA = Top_ShaleA[Top_ShaleA['VS']>0]
Top_ShaleA = Top_ShaleA[Top_ShaleA['RHOB']>0]

print(Top_ShaleA['VP'].mean())
print(Top_ShaleA['VS'].mean())
print(Top_ShaleA['RHOB'].mean())
```
